```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array([0, 6])
y = np.array([0, 250])

plt.plot(x, y)
plt.show()
```

```python
#plot without the line

x = np.array([0, 6])
y = np.array([0, 250])

plt.plot(x, y ,'o')
plt.show()
```

```python
#multiple points

x = np.array([5,10,15,20,25]) #by default: 0,1,2,3,4,5......
y = np.array([1,10,15,7,23])

plt.plot(x, y)
plt.show()
```

```python
#different markers

#x axis by default
y = np.array([10,20,30,40,50])
plt.plot(y, marker ='D') #changes marker design
#plt.plot(y,'*:r') #changes color
plt.show()

#all types of markers
#'o'		'*'		'.'		','	Pixel	'x'	X	'X'	X (filled) '+'	Plus	'P'	Plus (filled)	's'	Square	'D'	Diamond	
#'d'	Diamond (thin)	'p'	Pentagon	'H'	Hexagon	'h'	Hexagon	'v'	Triangle Down	'^'	Triangle Up	'<'	Triangle Left	
#'>'	Triangle Right	'1'	Tri Down	'2'	Tri Up	'3'	Tri Left	'4'	Tri Right	'|'	Vline	'_'	hline
```

```python
#line reference
#'-'	Solid line	':'	Dotted line	'--'	Dashed line	 '-.'	Dashed/dotted line

#color reference
#'r'	Red	'g'	Green	'b'	Blue	'c'	Cyan	'm'	Magenta	'y'	Yellow	'k'	Black	'w'	White

x = np.array([0, 6])
y = np.array([0, 250])

plt.plot(x, y ,'o:c') #changes color as well as line style
plt.show()
```

```python
#marker styles
ypoints = np.array([3, 8, 1, 10])
#ms : marker size, mec: marker color, mfc: marker face color
plt.plot(ypoints, marker = 'o', ms = 20, mec = 'r', mfc ='red')
plt.show()
```

```python
#line styles
ypoints = np.array([3, 8, 1, 10])
#line styles: 'solid' (default)	'-'	'dotted'	':'	'dashed'	'--'	'dashdot'	'-.'	'None'	'' or ' '	

plt.plot(ypoints, linestyle = 'dotted', color = 'r', linewidth = '4.0')

plt.show()
```

```python
#multiple lines
y1 = np.array([3, 8, 1, 10])
y2 = np.array([6, 2, 7, 11])

plt.plot(y1, color = 'hotpink')
plt.plot(y2, color = 'c')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])
x2 = np.array([0, 1, 2, 3])
y2 = np.array([6, 2, 7, 11])

#legends and titles
plt.title("Sports Watch Data" , loc = 'left') #Legal values are: 'left', 'right', and 'center'. Default value is 'center'.
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")


plt.plot(x1, y1, x2, y2)
plt.show()
```

```python
#grid 
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.plot(x, y)

#plt.grid() plt.grid(axis = 'x') plt.grid(axis = 'y')
plt.grid(color = 'green', linestyle = '--', linewidth = 0.5)


plt.show()
```

```python
#Draw 2 plots on top of each other:
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)

plt.show()
```

```python
#scatter diagram

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y, color = 'hotpink')
plt.show()

#color each dot
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array(["red","green","blue","yellow","pink","black","orange","purple","beige","brown","gray","cyan","magenta"])

plt.scatter(x, y, c=colors)

plt.show()
```

```python
#include colormap and change sizes

x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
colors = np.array([0, 10, 20, 30, 40, 45, 50, 55, 60, 70, 80, 90, 100])
sizes = np.array([20,50,100,200,500,1000,60,90,10,300,600,800,75])

plt.scatter(x, y, c=colors, cmap='viridis',s=sizes)

plt.colorbar()

plt.show()
```

```python
#bar graph

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y,color = 'hotpink',width = 0.5) #barh for horizontal and use height instead of width
plt.show()
plt.barh(x,y,color = 'hotpink',height = 0.5)
plt.show()
```

```python
#histogram

import matplotlib.pyplot as plt
import numpy as np

x = np.random.normal(170, 10, 250)
#print(x)
plt.hist(x,color ="darkblue")
plt.show() 
```

```python
#piechart with labels

import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
#mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
plt.pie(y, labels = ["Apples", "Bananas", "Cherries", "Dates"])
#plt.pie(y, labels = mylabels)
plt.show() 
```

```python
#startangle #explode #shadow
#bydefault it starts from X-axis but we can change it
#using explode we can make each wedge to stand out
y = np.array([35, 25, 25, 15])
plt.pie(y, labels = ["Apples", "Bananas", "Cherries", "Dates"],startangle = 180, explode = [0.2, 0, 0, 0],shadow =True)
plt.show() 
```

```python
#colors and legends
y = np.array([35, 25, 25, 15])
mycolors = ["black", "red", "b", "yellow"]
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplod = [0.2, 0, 0, 0]
plt.pie(y, labels = mylabels ,startangle = 90, explode = myexplod, colors = mycolors)
plt.legend(title = 'Fruits')
plt.show() 

```
