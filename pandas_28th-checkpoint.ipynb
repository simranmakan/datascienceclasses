```python
import requests as r
from bs4 import BeautifulSoup

url = 'https://www.w3schools.com/python/default.asp'
res = r.get(url)
soup = BeautifulSoup(res.text, 'html.parser')

# Get the left menu container
menu = soup.find('div', id='leftmenuinnerinner')

# Find all anchor tags in that menu
links = menu.find_all('a')

# Print each title + full link
for link in links:
    title = link.text.strip()
    href = link.get('href')
    
    # Handle any missing hrefs (just in case)
    if href:
        full_url = f"https://www.w3schools.com{href}"
        print(f"{title}  {full_url}")
```
